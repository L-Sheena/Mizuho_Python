```python
import requests
```

```python
url = "https://ia800302.us.archive.org/10/items/draculabr00stokuoft/draculabr00stokuoft_djvu.txt"
```

```python
data = requests.get(url).text
```

```python
data.lower().split("\n")
```

```python
newdata = []
```

```python
for eachL in data.lower().split("\n"):
    if eachL is not '':
        newdata.append(eachL)
```

```python
newdata
```

```python
words ={}
```

```python
for each in newdata:
    for word in each.split():
        if word not in words:
            words[word] = 1
        else:
            words[word] = words[word] +1
```
